#### 5. Handling Outliers

### IQR method

```python
# Calculate the IQR for Horsepower
Q1 = np.percentile(df.Price, 25)
Q3 = np.percentile(df.Price, 75)
IQR = Q3 - Q1

# Set a threshold for identifying outliers (e.g., values outside [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR])
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df.Price < lower_bound) | (df.Price > upper_bound)]
```

```python
print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")
```

#### Capping Outliers of MSRP using IQR Method

```python
# Cap Extreme values identified from Outliers
def cap_outliers(data, lower_threshold, upper_threshold):
    capped_data = np.clip(data, lower_threshold, upper_threshold)
    return capped_data

# Cap the outliers
capped_data = cap_outliers(df.Price, lower_bound, upper_bound)
capped_data

df.loc[:,'Price'] = capped_data
```

### 6. Apply Machine Learning Algorithm

```python
# Select only numeric type of columns by their names as:
x = df[['Battery','Processor','RAM(MB)','Internal_storage(GB)','Rear_camera','Front_camera','Wi-Fi','Bluetooth','GPS','Number of SIMs','3G','4G/ LTE','Resolution_y','Resolution_x']]
x
```

```python
# Select only numeric type of columns by their names as:
y = df[['Price']]
y
```

### Import Machine Leaning Libraries and Split Dataset

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
```

### Create, Fit, Predict and Evaluate Model
### 1. Multiple Linear Regression

```python
# Create a linear regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# Fit the model to the training data
lr.fit(X_train, y_train)

# Make predictions on the test data
y_pred = lr.predict(X_test)

# Evaluate the model
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(f"R-squared (Coefficient of Determination): {round(r2,2)}")
```

### 2. Decision Trees

```python
from sklearn.tree import DecisionTreeRegressor
# Create a Decision Tree model
dt = DecisionTreeRegressor(max_depth = 3)

# Fit the model to the training data
dt.fit(X_train, y_train)

# Make predictions on the test data
y_pred = dt.predict(X_test)

# Evaluate the model
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(f"R-squared (Coefficient of Determination): {round(r2,2)}")
```

### 3. Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

# Create a linear regression model
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model to the training data
rf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf.predict(X_test)

# Evaluate the model
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(f"R-squared (Coefficient of Determination): {round(r2,2)}")
```

### 4. SVR

```python
from sklearn.svm import SVR

# Create a SVR regression model
model = SVR(kernel='linear')

# Fit the model to the training data
rf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf.predict(X_test)

# Evaluate the model
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(f"R-squared (Coefficient of Determination): {round(r2,2)}")
```
